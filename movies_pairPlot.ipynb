```python
import seaborn as sns
import pandas as pd # `pd` is a common alias for `pandas`

df = pd.read_csv('tmdb_5000_movies.csv')  # df is a common name for `DataFrame`
df_relevant = df[
    [
        'title', 'budget', 'revenue', 'popularity', 
        'release_date', 'runtime', 'vote_average', 'vote_count'
     ]
]
df_relevant = df_relevant.loc[
    (df_relevant['budget'] > 0) &
    (df_relevant['revenue'] > 0) &
    (df_relevant['vote_count'] > 100) 
    ].copy()

# this cell looks scary, you'll learn about it later (by yourself)
df_relevant['release_date'] = pd.to_datetime(df_relevant['release_date'])
df_recent = df_relevant.set_index('release_date').sort_index().loc['1990':].copy()
df_recent
sns.pairplot(df_recent)
```
